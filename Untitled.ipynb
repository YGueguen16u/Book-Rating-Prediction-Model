```python
with open("books.csv", 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Identify the line numbers with the incorrect number of commas (assuming 11 commas for 12 columns)
incorrect_lines = [(index, line) for index, line in enumerate(lines) if line.count(',') != 11]

for index, line in incorrect_lines:
    print(f"Ligne {index}: {line}")
```

### The Author's  column error

**The Issue 🚩**

We observe that there are several lines with this same problem. Fortunately, the problematic field is consistently the same - the authors field. Specifically, for these lines, the authors field contains a ',', which splits the field into two separate parts, thereby creating an additional field.

**The Solution 🔧**

* ***Step 1:*** Scan all the lines. Check if there are more fields than anticipated for each line.
* ***Step 2:*** If there's a discrepancy, merge the author's field from that line and the subsequent field into one.
* ***Step 3:*** Save the modified data into a new file, preserving the original books file.
* ***Step 4:*** Load and display the dataset as a pandas dataframe.

```python
fixed_rows = []
with open("books.csv", "r") as file:
    lines = file.readlines()
    for line in lines:
        # If the line has more fields than expected
        if line.count(",") > 11:
            parts = line.split(",")
            # Merging author segments to remove the excess comma
            parts[2] = parts[2] + parts[3]
            # Removing the excess segment
            del parts[3]
            fixed_rows.append(",".join(parts))
        else:
            fixed_rows.append(line)

# Save to a new CSV file
with open("fixed_books.csv", "w") as file:
    file.writelines(fixed_rows)

books_df = pd.read_csv("fixed_books.csv", index_col="bookID")
```

#### Date Error Correction

**Objective:**

Automated correction of inaccurate date entries in our dataset.

**Methodology:**

* Identifying invalid dates, such as June 31st.
* Adjusting these entries by decrementing by one day to ensure a valid date.

**Added Features:**

* Avoiding Redundancies: Checks have been established to prevent row duplication when writing to the corrected CSV file.In an initial attempt, the new file created redundancies, resulting in a file that was ten times larger.
* Comparing CSV Files: A comparison routine has been integrated to ensure that our CSV files (original and corrected) align in terms of book identifiers (bookID). This step ensures the consistency and integrity of our data throughout the changes.

```python
from datetime import datetime, timedelta
```

```python
invalid_dates = ["6/31", "11/31"]  # June 31st and November 31st

invalid_book_ids = []

with open("fixed_books.csv", "r") as file:
    header = file.readline()  
    for line in file:
        parts = line.split(",")
        date = parts[10].strip()
        if any(invalid_date in date for invalid_date in invalid_dates):
            invalid_book_ids.append(parts[0])
            
corrected_rows = []

with open("fixed_books.csv", "r") as file:
    header = file.readline().strip()  # Strip helps remove any extra newline characters
    corrected_rows.append(header)  # Preserve the first row with column's features

    for line in file:
        parts = line.strip().split(",")  
        book_id = parts[0]
        
        if book_id in invalid_book_ids:
            # Extract date components
            month, day, year = map(int, parts[10].strip().split('/'))
            
            # Decrement the day
            day -= 1
            
            # Handle cases where day becomes 0 (roll over to the last day of the previous month)
            if day == 0:
                month -= 1
                if month == 0:
                    month = 12
                    year -= 1
                last_day_of_prev_month = (datetime(year, month + 1, 1) - timedelta(days=1)).day
                day = last_day_of_prev_month
            
            # Reassign corrected date to the list
            parts[10] = f"{month:02}/{day:02}/{year}"
            corrected_rows.append(",".join(parts))
        else:
            corrected_rows.append(line.strip())

# Save to the same CSV file
with open("fixed_books_2.csv", "w") as file:
    file.write("\n".join(corrected_rows))

# Reload the dataframe
books_df = pd.read_csv("fixed_books_2.csv")

# Check if two CSVs have the same bookIDs
def csvs_same_book_ids(csv1, csv2):
    df1 = pd.read_csv(csv1)
    df2 = pd.read_csv(csv2)
    ids_df1 = set(df1['bookID'].tolist())
    ids_df2 = set(df2['bookID'].tolist())
    return ids_df1 == ids_df2

# Test
same_csv = csvs_same_book_ids("fixed_books.csv", "fixed_books_2.csv")
print("Les deux CSV ont-ils les mêmes bookIDs?", same_csv)
```

**Redudant author's name**

While exploring the dataset in this Jupyter Notebook, we observe that certain books have the same author listed multiple times within the authors_list. To enhance the clarity of our data, we'll pinpoint which books have this characteristic and ensure each author is listed only once. To achieve this, we'll employ a string similarity technique leveraging the Levenshtein distance - a method I discovered online. 


**Levenshtein Distance - A Brief Overview**

* **Definition**: Measures the difference between two strings, often called the "edit distance".

* **Purpose**: Determines how many edits (insertions, deletions, or substitutions) are required to transform one string into another.

* **Applications**: Widely used in:

    * Spell-checking
    * DNA sequence analysis
    * Search optimizations

* **Example**:
For "kitten" and "sitting", the Levenshtein distance is 3.
    * kitten → sitten (substitution of "s" for "k")
    * sitten → sittin (substitution of "i" for "e")
    * sittin → sitting (insertion of "g" at the end)


```python
def levenshtein_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for index2, char2 in enumerate(s2):
        new_distances = [index2 + 1]
        for index1, char1 in enumerate(s1):
            if char1 == char2:
                new_distances.append(distances[index1])
            else:
                new_distances.append(1 + min((distances[index1], distances[index1 + 1], new_distances[-1])))
        distances = new_distances

    return distances[-1]

def similar_names(name1, name2, threshold=0.84):
    """
    Compare two names and determine if they are similar using the Levenshtein distance.

    :param name1: First name to compare.
    :param name2: Second name to compare.
    :param threshold: Similarity threshold between 0 and 1, where 1 indicates a perfect match.
    :return: True if the names are considered similar, otherwise False.
    """
    
    # Calculate the similarity ratio based on the Levenshtein distance.
    max_len = max(len(name1), len(name2))
    if max_len == 0:
        return True

    similarity = 1 - (levenshtein_distance(name1, name2) / max_len)
    
    return similarity > threshold

def has_similar_authors(authors_list):
    """
    Checks if a list of authors contains similar authors.

    :param authors_list: List of author names.
    :return: True if similar authors are found, otherwise False.
    """

    for i in range(len(authors_list)):
        for j in range(i+1, len(authors_list)):
            if similar_names(authors_list[i], authors_list[j]):
                return True
    return False

books_df['has_similar'] = books_df['authors_list'].apply(has_similar_authors)

similar_authors_df = books_df[books_df['has_similar'] == True]

print(len(similar_authors_df))
similar_authors_df
```

**Affected Names Analysis**

We aim to determine if multiple names are impacted by this method.thod.

```python
def count_similar_authors(authors_list):
    """
    Counts how many authors in the list are similar.

    :param authors_list: List of author names.
    :return: Number of similar authors in the list.
    """
    count = 0
    for i in range(len(authors_list)):
        for j in range(i+1, len(authors_list)):
            if similar_names(authors_list[i], authors_list[j]):
                count += 1
    return count

books_df['authors_list'] = books_df['authors'].str.split('/')

books_df['similar_authors_count'] = books_df['authors_list'].apply(count_similar_authors)

similar_authors_df = books_df[books_df['similar_authors_count'] > 0]

print(len(similar_authors_df))
similar_authors_df
```

**Names Overview**

Below, we showcase the authors in question for each book.

```python
def get_similar_authors_pairs(authors_list):
    """
    Renvoie les paires d'auteurs qui sont similaires dans la liste.

    :param authors_list: Liste des noms d'auteurs.
    :return: Liste de tuples contenant des paires d'auteurs similaires.
    """
    similar_pairs = []
    for i in range(len(authors_list)):
        for j in range(i+1, len(authors_list)):
            if similar_names(authors_list[i], authors_list[j]):
                similar_pairs.append((authors_list[i], authors_list[j]))
    return similar_pairs

books_df['similar_authors_pairs'] = books_df['authors_list'].apply(get_similar_authors_pairs)
books_df['similar_authors_count'] = books_df['similar_authors_pairs'].apply(len)

similar_authors_df = books_df[books_df['similar_authors_count'] > 0]
print(len(similar_authors_df))
similar_authors_df[['title', 'similar_authors_count', 'similar_authors_pairs']]
```

**Removal of Redundant Names**

In this section, we'll remove all redundant names from the authors_list

```python
#before removing 
print ("Before removal : ", books_df.loc[5285, "authors_list"])
```

```python
def remove_one_redundant_author(authors_list, pairs):
    """
    Removes one occurrence of each redundant author from the author list.

    :param authors_list: List of author names.
    :param pairs: List of similar author pairs.
    :return: List of authors after removing one occurrence of each redundant author.
   
    """
    for _, redundant_author in pairs: 
        if redundant_author in authors_list:
            authors_list.remove(redundant_author)
    return authors_list

books_df['cleaned_authors_list'] = books_df.apply(lambda row: remove_one_redundant_author(row['authors_list'].copy(), row['similar_authors_pairs']), axis=1)

books_df['authors_list'] = books_df['cleaned_authors_list']
books_df.drop('cleaned_authors_list', axis=1, inplace=True)

print ("after removal : ", books_df.loc[5285, "authors_list"])
```

### Verify that There Are No Aberrant Values in the Number of Pages

In this section, we will focus on ensuring the integrity and reliability of our book dataset. The following tasks will be performed:

* Identify Outliers: Examination of the dataset to identify and manage extreme values, such as books with only 1 page or an unrealistic 10,000 pages.

* Verify Positive Values: Validation to ensure that all numerical fields, particularly the number of pages, contain positive values.

**Column Renaming for Clarity**

We notice that the column name for the number of pages has extra spaces, which could lead to errors or confusion in future code. To streamline our DataFrame and make it more readable, we'll rename this column to `num_pages`.

```python
# Renaming for clarity
books_df.rename(columns={'  num_pages': 'num_pages'}, inplace=True)

# Identifying outlier values
outliers = books_df[(books_df['num_pages'] < 2) | (books_df['num_pages'] > 10000)]
if outliers.shape[0] > 0:
    print(f"{outliers.shape[0]} outliers detected.")
    print(outliers[['bookID', 'title', 'num_pages']])
else:
    print("No outliers detected.")

# Checking for non-positive values
non_positive_values = books_df[books_df['num_pages'] < 0]
if non_positive_values.shape[0] > 0:
    print(f"{non_positive_values.shape[0]} non-positive values detected.")
    print(non_positive_values[['bookID', 'title', 'num_pages']])
else:
    print("All values are positive.")
```

In our dataset, the num_pages column has some aberrant values—some books have either too few or too many pages. To handle these outliers, one approach could be to impute these missing or incorrect values based on other features in the dataset.

A correlation matrix can help us understand the relationship between num_pages and other numeric features. If a strong correlation exists, we can use that feature to estimate the missing or aberrant num_pages values.

Note: Correlation does not imply causation; it only indicates the strength and direction of a relationship between two variables.


```python
correlation_matrix = books_df.corr()
print(correlation_matrix['num_pages'])
```

After conducting a correlation analysis between the num_pages column and other numerical features in our dataset, we find that num_pages has no significant correlation with any other column. The correlation coefficients are all near zero, with the highest being only 0.150809 with average_rating.

Given this lack of correlation, imputing or correcting the outliers in the num_pages column based on other features is not advisable. Therefore, it may be prudent to remove entries with aberrant num_pages values from the dataset to maintain data integrity.

**Frequency Encoding of Categorical Variables**

In the given code snippet, Frequency Encoding is applied to the 'publisher' and 'second_language_code' columns of the books_df DataFrame. Here's a quick breakdown:

* **Calculate Frequency**: For each unique value in the categorical columns, the frequency of occurrence is calculated as a proportion of the total dataset size.

    * freq_publisher = books_df['publisher'].value_counts() / len(books_df)
    * freq_language_code = books_df['second_language_code'].value_counts() / len(books_df)


* **Map Frequencies**: These calculated frequencies are then mapped back to new columns ('publisher_encoded' and 'language_code_encoded') in the DataFrame.

    * books_df['publisher_encoded'] = books_df['publisher'].map(freq_publisher)
    * books_df['language_code_encoded'] = books_df['second_language_code'].map(freq_language_code)

The approach ensures that the same frequency value is assigned to all instances of the same category. This method is particularly useful when dealing with categorical variables that have a large number of categories, as it avoids increasing the data dimensionality.

```python
# Calculate the frequency of each 'publisher' and 'language_code'
freq_publisher = books_df['publisher'].value_counts() / len(books_df)
freq_language_code = books_df['second_language_code'].value_counts() / len(books_df)

# Map the frequencies back to new columns
books_df['publisher_encoded'] = books_df['publisher'].map(freq_publisher)
books_df['language_code_encoded'] = books_df['second_language_code'].map(freq_language_code)
```

**Metrics Overview**

* **Mean Squared Error (MSE):** Measures the average of the squares of the errors, which means it considers both the magnitude and direction of the prediction errors.


* **Root Mean Squared Error (RMSE):** The square root of the MSE, providing error metric in the same unit as the output.


* **Mean Absolute Error (MAE):** Represents the average absolute difference between the observed actual values and the values predicted by the model.


* **R^2 Score:** Represents the proportion of the variance for the dependent variable that is explained by the independent variables in the model. It ranges from -1 to 1.

### Comparative Analysis of Regression Models

Different regression models have their unique strengths, weaknesses, and assumptions. The metrics obtained from the Linear Regression, Random Forest, AdaBoost, and Decision Tree models shed light on their performance, but it's crucial to understand why these models could produce varying results:


**Linear Regression**: This model assumes a linear relationship between the independent and dependent variables. If this assumption doesn't hold, the model may not capture the underlying pattern well, which seems to be the case given its low R^2 score.


**Random Forest**: This ensemble model combines multiple decision trees and generally does a better job at capturing non-linear relationships. Its higher R^2 and lower MSE suggest it's performing better than the linear model.


**AdaBoost**: Another ensemble method, but it works by adjusting the weights of observations. It lies between the Linear Regression and Random Forest in terms of performance, perhaps because it focuses more on the difficult to predict observations.


**Decision Tree**: Generally, a single decision tree is prone to overfitting, and its poor metrics here could be an indicator 
of that phenomenon. It performs significantly worse, as evidenced by its negative R^2 and higher MSE.



**Why Do These Differences Exist?**


**Model Complexity**: Linear Regression is a simpler model compared to Random Forest and AdaBoost. The latter can model complex relationships but might require more tuning.


**Non-Linearity:** Decision Trees and Random Forests can capture non-linear relationships, which Linear Regression cannot.


**Bias-Variance Tradeoff:** Linear models generally have high bias but low variance. On the other hand, models like Random Forest have low bias but high variance. AdaBoost tries to balance the two but requires careful tuning.


**Data Sensitivity:** Some models are more sensitive to outliers and feature scale. AdaBoost is one such model, which is why it may not perform as well as Random Forest in some cases.


**Overfitting:** Decision Trees are prone to overfitting unless properly pruned, which may explain its poor performance.


By comparing these metrics and understanding the characteristics of each model, one can make a more informed decision when choosing a model for a specific task.
